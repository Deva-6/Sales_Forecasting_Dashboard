Project: Sales Forecasting Dashboard

Goal: Predict next 90 days of sales from historical data

Method: Average of last 30 days used for forecasting

Output: Line chart dashboard + 30-day prediction table

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dates = pd.date_range('2024-01-01', '2024-12-31')
sales = [1000 + (i*2) + (50 if d.dayofweek >= 5 else 0) for i, d in enumerate(dates)]

df = pd.DataFrame({'Date': dates, 'Sales': sales})
df.head()
```

```python
last_30_avg = df['Sales'].tail(30).mean()
future_dates = pd.date_range('2025-01-01', '2025-03-31')
forecast = [last_30_avg] * 90
forecast_df = pd.DataFrame({'Date': future_dates, 'Forecast': forecast})
forecast_df.head()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(df['Date'], df['Sales'], label='Historical', color='blue')
plt.plot(forecast_df['Date'], forecast_df['Forecast'], label='Forecast', color='red', linestyle='--')
plt.title('Sales Forecast')
plt.xlabel('Date')
plt.ylabel('Sales ($)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print(forecast_df.head(30))
```

This project analyzes historical sales data to predict future trends.
Using simple moving average method based on last 30 days of sales.
The dashboard displays past sales with 90-day forecast in line chart.
Output includes prediction table and downloadable CSV for business planning.
